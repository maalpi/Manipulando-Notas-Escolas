## Descrição
Todos os anos, os estudantes americanos do ensino médio fazem SATs, que são testes padronizados destinados a medir habilidades de alfabetização, numeramento e escrita. Existem três seções - leitura, matemática e redação, cada uma com pontuação máxima de 800 pontos. Esses testes são extremamente importantes para estudantes e faculdades, pois desempenham um papel fundamental no processo de admissão.

A análise do desempenho das escolas é importante para uma variedade de partes interessadas, incluindo profissionais da política e da educação, investigadores, governo e até mesmo pais que consideram que escola os seus filhos devem frequentar.

 - Os melhores resultados em matemática são pelo menos 80% da *pontuação máxima possível de 800* em matemática.

 > Segunda Tarefa
 - Quais são as 10 escolas com melhor desempenho com base nas pontuações combinadas do SAT?
  - Basicamente tem que criar um dataframe com as 10 escolas que tiverem melhores pontuacoes

  > Terceira Tarefa
  - Qual bairro tem o maior desvio padrão na pontuação combinada do SAT?
    - “borough” — o nome do bairro de Nova York com o maior desvio padrão de “total_SAT”.
    - “num_schools” — o número de escolas no bairro.
    - “average_SAT” — a média de “total_SAT”.
    - “std_SAT” — o desvio padrão de “total_SAT”.



```python
# Re-run this cell
import pandas as pd

# Read in the data
schools = pd.read_csv('/content/sample_data/schools.csv')

# Preview the data
schools.head()
```

```python
# Escolas com a maior pontuação e matematica acima do 640
best_math_schools = schools[schools["average_math"] >= 640]
best_math_schools = best_math_schools.sort_values(["average_math"], ascending=False)
best_math_schools = best_math_schools[["school_name","average_math"]]
best_math_schools
```

```python
# Filtrando as escolas que possuem a maior perfomace, soma das notas em todos os testes
top_10_schools = pd.DataFrame(schools["school_name"])
top_10_schools["total_SAT"] = schools["average_math"] + schools["average_reading"] + schools["average_writing"]
top_10_schools = top_10_schools.sort_values("total_SAT",ascending=False).head(10)
top_10_schools
```

```python
# Calculando desvio padrão, numero de escolas e o valor medio da coluna
# No final eu crio um dataframe so com a escola que possui o maior desvio padrao e suas informacoes
schools["total_SAT"] = schools["average_math"] + schools["average_reading"] + schools["average_writing"]
boroughs = schools.groupby('borough')['total_SAT'].agg(['count', 'mean', 'std']).round(2)
value = boroughs['std'].max()
largest_std_dev = boroughs[boroughs["std"] == value]
largest_std_dev = largest_std_dev.rename(columns={"count":"num_schools", "mean":"average_SAT", "std":"std_SAT"})
largest_std_dev
```
